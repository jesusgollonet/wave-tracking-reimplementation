```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
print('hii')
```

# Normal distribution
skewness and kurtosis are close to 0 given it's a normal distribution

```python
data = np.random.normal(loc=0, scale=1, size=1000)
plt.scatter(range(1000), data)
mean = np.mean(data)
plt.axhline(mean, color='red')
variance = np.var(data)
plt.axhline(mean + variance, color='green')
plt.axhline(mean - variance, color='green')
skewness = stats.skew(data)
plt.axhline(mean + skewness, color='purple')
kurtosis = stats.kurtosis(data)
plt.axhline(mean + kurtosis, color='orange')
```

```python
# Skewed Distribution: Log-normal
skewed_data = np.random.lognormal(mean=0, sigma=1, size=1000)
plt.scatter(range(1000), skewed_data)
mean = np.mean(skewed_data)
plt.axhline(mean, color='red')
variance = np.var(skewed_data)
plt.axhline(mean + variance, color='green')
skewness = stats.skew(skewed_data)
plt.axhline(mean + skewness, color='purple')
kurtosis = stats.kurtosis(skewed_data)
plt.axhline(mean + kurtosis, color='orange')
```
