```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv.imread('media/t1.png', cv.IMREAD_GRAYSCALE)
img2 = cv.imread('media/t2.png', cv.IMREAD_GRAYSCALE)

contour1, _ = cv.findContours(img1, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
contour2, _ = cv.findContours(img2, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

intersection = cv.bitwise_and(img2, img1)


fig, ax = plt.subplots(1, 3)
ax[0].imshow(img1, cmap='gray')
ax[1].imshow(img2, cmap='gray')
ax[2].imshow(intersection, cmap='gray')
```
