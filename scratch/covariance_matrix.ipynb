```python
# generate a set of 2d points and plot them
import numpy as np
import matplotlib.pyplot as plt

# generate a set of 2d points
n = 100
# generate a line of x values with some random noise
x = np.linspace(1, 0, n) + np.random.rand(n)/2 
y = np.linspace(0, 1, n) + np.random.rand(n)/2

data = np.column_stack((x, y))
# plot the points
plt.scatter(x, y)
plt.show()
```

```python
cov = np.cov(x, y)
eigvals, eigvecs = np.linalg.eig(cov)
for eigval, eigvec in zip(eigvals, eigvecs.T):
    print(eigval, eigvec)
    plt.quiver(*data.mean(axis=0), *eigvec * 2 * np.sqrt(eigval), angles='xy', scale_units='xy', scale=1, color='red')
plt.scatter(x, y)
plt.show()
```
